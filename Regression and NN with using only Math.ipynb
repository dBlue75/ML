```python
from numpy import random
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
srf_df2=pd.DataFrame(columns=['X','Y'])
for x in range (1,10001):
    srf_x2=x/1000
    srf_y2 = math.log(srf_x2)+0.3*math.sin(srf_x2)+0.1*math.sin(srf_x2**2)+np.random.normal(0,0.1)
    srf_df2.loc[x]=([srf_x2,srf_y2])
    
input_X=(np.array(srf_df2['X'])).reshape(len(srf_df2),1)
output_y=(np.array(srf_df2['Y'])).reshape(len(srf_df2),1)
```

```python
#learning rate 
alpha = 0.005
present_cost=0
previous_cost=1000000000
```

```python
def sigmoid(value):
    a=(1/(1+np.exp(-value)))
    return a 
```

```python
def tanh(x):
    t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    return t
```

```python
def deri_sigmoid(value):
    a=np.multiply(sigmoid(value),np.subtract(np.ones(value.shape),sigmoid(value)))
    return a
```

```python
# a is the activation value of the last layer
def cost(a):
    cost_=(0.5)*np.matmul(np.transpose(np.subtract(a,output_y)),np.subtract(a,output_y))
    return cost_
```

```python
def sub_cost(a):
    cost_=np.subtract(a,output_y)
    return cost_
```

```python
W1= random.rand(10,len(input_X))
W2=random.rand(10,10)
W3=random.rand(len(output_y),10)
```

```python
#FOREWARDPROP
def for_prop():
    global W1,W2,W3,a1,a2,a3
    
    a1=sigmoid(np.matmul(W1,input_X))

    a2=sigmoid(np.matmul(W2,a1))

    a3=np.matmul(W3,a2)
    
    return(cost(a3))
```

```python
#BACKPROP
def back_prop():
    global W1,W2,W3,a1,a2,a3
    
    W3_error=np.multiply(sub_cost(a3),(np.matmul(W3,a2)))
    W3=W3-(alpha*(np.matmul(W3_error,np.transpose(a2))))

    W2_error=np.multiply(np.matmul(np.transpose(W3),W3_error),deri_sigmoid(np.matmul(W2,a1)))
    W2=W2-(alpha*(np.matmul(W2_error,np.transpose(a1))))

    W1_error=np.multiply(np.matmul(np.transpose(W2),W2_error),deri_sigmoid(np.matmul(W1,input_X)))
    W1=W1-(alpha*(np.matmul(W1_error,np.transpose(input_X))))

    #previous_cost = present_cost
    #present_cost =  cost(a3)
```

```python
ccost=[]
while True :
    present_cost=for_prop()
    back_prop()
    
    ccost.append(present_cost[0][0])
    left_cost=(previous_cost-present_cost)
    if left_cost < 0.001:
        break
    previous_cost=present_cost
```

```python
x = np.linspace(ccost[-1],ccost[0],len(ccost))
y = ccost
fig = plt.figure(figsize=(20,10))
plt.plot(x, y, '-r')
#plt.plot(srf_df2['X'],srf_df2['Y'])
plt.grid()
plt.show()
```
